**Downloading youre SDXL model Just give the dowload URL**

```python
# Step 3: Download the model
import os
import urllib.request
from tqdm import tqdm
import requests

%cd /content/stable-diffusion-webui-reForge/models/Stable-diffusion

def download_with_progress(url, filename):
    response = requests.get(url, stream=True)
    total_size = int(response.headers.get('content-length', 0))
    with open(filename, 'wb') as f, tqdm(total=total_size, unit='B', unit_scale=True, desc="Downloading") as pbar:
        for chunk in response.iter_content(chunk_size=1024):
            f.write(chunk)
            pbar.update(len(chunk))

download_with_progress("https://civitai.com/api/download/models/1183765?type=Model&format=SafeTensor&size=pruned&fp=fp16", "WAI-NSFW-illustrious-SDXL.safetensors")
%cd /content/stable-diffusion-webui-reForge
#you can use !wget
```
